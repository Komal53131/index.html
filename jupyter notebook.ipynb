```python
import pandas as pd

# Load the dataset
df = pd.read_csv('orders.csv')

# Convert order_date to datetime objects for time-based analysis
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)

# --- Q1: Total Revenue Generated ---
total_revenue = df['total_amount'].sum()
print(f"Total Revenue: ${total_revenue:,.2f}")

# --- Q2: Top 5 Most Popular Restaurants (by order count) ---
top_restaurants = df['restaurant_name'].value_counts().head(5)
print("\nTop 5 Restaurants by Order Count:")
print(top_restaurants)

# --- Q3: Average Order Value (AOV) ---
aov = df['total_amount'].mean()
print(f"\nAverage Order Value: ${aov:.2f}")

# --- Q4: Monthly Order Trends ---
df['month'] = df['order_date'].dt.strftime('%B')
monthly_orders = df['month'].value_counts()
print("\nOrders by Month:")
print(monthly_orders)
```
